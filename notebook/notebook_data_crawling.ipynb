# Using RapidAPI

```python
from dotenv import load_dotenv
import os

load_dotenv()

api_key= os.getenv("API_KEY")
api_host = os.getenv("API_HOST")
```

```python
import requests

url = "https://zillow56.p.rapidapi.com/search"

querystring = {"location":"houston",
			   "output":"json",
			   "status":"forSale",
			   "sortSelection":"priorityscore",
			   "listing_type":"by_agent",
			   "doz":"any"}

headers = {
	"x-rapidapi-key": api_key,
	"x-rapidapi-host": api_host
}

# Initialize an empty DataFrame
df = pd.DataFrame()

# Set the initial page number
i = 0

# Loop to get data from API
while True:
    try:
        # Update the querystring with the current page number
        querystring["page"] = i

        # Make the GET request to the API
        response = requests.get(url, headers=headers, params=querystring)

        # Check if the response contains data
        if response.status_code != 200 or 'results' not in response.json():
            break

        # Normalize the JSON response to a DataFrame
        temp = pd.json_normalize(response.json()['results'])

        # Concatenate the new data with the existing DataFrame
        df = pd.concat([df, temp], ignore_index=True)

        # Increment the page number
        i += 1

    except Exception as e:
        print(f"An error occurred: {e}")
        break

# Output the resulting DataFrame
print(df)
```

```python
today = datetime.datetime.now().strftime(format="%Y%m%d")
df.to_csv(f'../data/data_{today}.csv', index=False)
```
